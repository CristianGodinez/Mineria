####Conceptos


##### La **Minería de Datos (Data Mining)** es el proceso de descubrir patrones, tendencias y conocimientos significativos a partir de grandes conjuntos de datos. Es una etapa crucial dentro de la Ciencia de Datos, que se enfoca en la aplicación de algoritmos y técnicas para extraer información valiosa que no es evidente a simple vista.

Los objetivos principales de la minería de datos incluyen:

*   **Predicción:** Construir modelos para predecir valores futuros o comportamientos. (Ej: predicción de ventas, identificación de riesgo de abandono de clientes).
*   **Clasificación:** Asignar elementos a categorías predefinidas. (Ej: detección de fraude, clasificación de correos electrónicos como spam).
*   **Agrupamiento (Clustering):** Identificar grupos de datos que comparten características similares sin conocimiento previo de las categorías. (Ej: segmentación de clientes).
*   **Asociación:** Descubrir relaciones entre variables en grandes bases de datos. (Ej: reglas de asociación como 'si compra X, probablemente comprará Y').
*   **Detección de Anomalías:** Identificar eventos raros o atípicos que pueden indicar problemas o nuevas oportunidades. (Ej: detección de intrusiones en redes, anomalías en transacciones bancarias).

La minería de datos utiliza una combinación de técnicas estadísticas, inteligencia artificial y machine learning para llevar a cabo estos procesos.


####Ciencia de datos

La Ciencia de Datos es un campo interdisciplinario que utiliza métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento o _insights_ de datos en diversas formas, tanto estructuradas como no estructuradas. Combina elementos de estadística, matemáticas, informática y conocimiento del dominio para analizar datos y tomar decisiones basadas en ellos.

En esencia, la ciencia de datos se enfoca en:

*   **Recolección y preparación de datos:** Obtener datos de diversas fuentes y limpiarlos para asegurar su calidad y usabilidad.
*   **Análisis exploratorio de datos (EDA):** Entender las características principales de los datos, identificar patrones, detectar anomalías y probar hipótesis.
*   **Modelado:** Aplicar algoritmos de machine learning, estadística o inteligencia artificial para construir modelos predictivos o descriptivos.
*   **Comunicación de resultados:** Presentar los hallazgos de manera clara y efectiva a las partes interesadas, a menudo utilizando visualizaciones de datos.
